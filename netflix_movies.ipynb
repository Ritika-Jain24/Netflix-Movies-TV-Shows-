```python
df['title']=df['title'].str.strip()
df.head(2)
```

```python
df['date_added']= pd.to_datetime(df['date_added'].str.strip())
df['year_added']= df['date_added'].dt.year
df
```

```python
df['year_added'] = df['year_added'].fillna(0).astype(int)
df
```

```python
df['year_added']= df['year_added'].replace (0, np.nan)
df


```

```python
df['year_added'] = df['year_added'].astype('Int64')
df
```

```python
df.dropna(subset=['director', 'cast', 'country'],how = 'any',inplace = True)
```

```python
df['is_movie'] = df['type'].apply(lambda x: 1 if x == 'Movie' else 0)
```

```python
df['movie_duration'] = df.apply(
    lambda x: x['duration'] if x['type'] == 'Movie' else None,
    axis=1
)
df
```
